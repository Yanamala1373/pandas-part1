```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("C:\\Users\\yanam\\OneDrive\\Desktop\\data science\\MARKS.xlsx")
print(df)
```

```python
'''ISNULL() is used to know where is the null value (NaN) in cell give "TRUE" in that coloumn'''
print(df.isnull())
```

```python
''' now to handle the missing value we have some options'''
'''OPTION 1

dropna() will delete the entire row where we encounter the NULL value (NaN) in cell value'''
a=df.dropna()
print(a)
```

```python
''' OPTION 2

fillna() will give you to fill the required value in the place where we encounter the NaN value in cell value'''
 
d=df['TELUGU'].fillna(df['TELUGU'].var())#instead of mean we can run mode ,medain,std (standard deviation),var(variance)
print(d)
```

```python
import math
C=list(df.columns)
print(C)
for i in range(2,len(C)):
    df[C[i]].fillna(math.ceil(df[C[i]].mean()),inplace=True)
print(df)
```

```python
'''TILL NOW WE HAVE SEEN THE DEALING WITH MISSING VALUES SO NOw WE NEED TO HOW REMOVE DUPLICATE VALUES'''
df.duplicated()
#if there any it will give you true and we can correct it by function(here u.prabhas is get deleted since it is duplicated) 
df.drop_duplicates()
print(df)
```
